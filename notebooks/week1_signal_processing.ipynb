```python
import pandas as pd
```

```python
df = pd.read_csv('ai4i2020.csv')
df.head()
```

```python
print(df.shape)
```

```python
print(df.columns.tolist())
```

```python
print(df['Machine failure'].value_counts())
```

```python
sensor_cols = [
    'Air temperature [K]',
    'Process temperature [K]',
    'Rotational speed [rpm]',
    'Torque [Nm]',
    'Tool wear [min]'
]
```

```python
window_size = 10  

for col in sensor_cols:
    df[f'{col}_rolling_mean'] = df[col].rolling(window=window_size).mean()
    df[f'{col}_rolling_std'] = df[col].rolling(window=window_size).std()
    df[f'{col}_rolling_var'] = df[col].rolling(window=window_size).var()
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
print(df.shape)
```

```python
rolling_cols = [c for c in df.columns if 'rolling' in c]
print(rolling_cols)
df[rolling_cols].head()
```
